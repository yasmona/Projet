```python
from keras.layers import RepeatVector
from keras.layers import TimeDistributed
from keras.models import Sequential
from keras.layers import LSTM,Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
import numpy as np
import string
```

```python
import pandas as pd
pd.__version__
df = pd.read_csv('C:/tweet.txt', sep="\n",header=None)
df
```

```python
list_listword=[]#liste des listes des mots de chaque tweet
list_tweets=df[0].values.tolist()#liste des tweets
list_tweets
```

```python
import re
l=[]
for text in list_tweets:
    text = re.sub(r'http\S+', '', text)   # Remove URLs
    text = re.sub(r'—', ' ', text) 
    text = re.sub(r'@[a-zA-Z0-9_]+', '', text)  # Remove @ mentions
    text = text.strip(" ")   # Remove whitespace resulting from above
    text = re.sub(r' +', ' ', text)   # Remove redundant spaces
    l.append(text)
list_tweets=l
list_tweets
```

```python
print(list_tweets[0])
for i in range(0,len(list_tweets)):
  list_listword.append(list_tweets[i].split())
list_listword
```

```python


tokenizer = Tokenizer()
tokenizer.fit_on_texts(list_tweets)
sequences = tokenizer.texts_to_sequences(list_tweets)
vocab_size=len(tokenizer.word_index)
max_length=0
for i in range(len(sequences)):
  if(len(sequences[i])>max_length):
    max_length=len(sequences[i])
print("max",max_length)
print("seq=",sequences)

X =sequences[:33]
X = pad_sequences(X, maxlen=max_length, padding='post')
print("X=",X)
print(X.shape)
y =sequences[33:]
y = pad_sequences(y, maxlen=max_length, padding='post')
print("y=",y)
print(y.shape)
X=X.reshape(33, max_length, 1)
y=y.reshape(33, max_length, 1)
print(X)
print(y)
print('train_x shape:', X.shape)
print('train_y shape:', y.shape)
```

```python
model = Sequential()

model.add(LSTM(57, input_shape=(max_length, 1), return_sequences=True))
#model.add(LSTM(57, activation='relu'))
model.add(Dense(57, activation="relu"))
model.add(Dense(57, activation="relu"))
model.add(TimeDistributed(Dense(1)))
model.compile(loss='mean_squared_error', optimizer='adam')
print(model.summary())
```

```python
#model.fit(X, y, epochs=100, validation_split=0.2, verbose=1, batch_size=3)
model.fit(X, y, epochs=100, batch_size=1, verbose=2)
```

```python

test_input=X[7:9]
print(y[7:9])
print(test_input)
encoded = pad_sequences([test_input], maxlen = 57, truncating='pre')

result = model.predict(test_input, batch_size=3, verbose=0)
print(result)
```
