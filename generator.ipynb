```python
from keras.layers import RepeatVector
from keras.layers import TimeDistributed
from keras.models import Sequential
from keras.layers import LSTM,Dense,Input,ConvLSTM2D,Reshape,Activation,Lambda,Softmax
from tensorflow.keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
import numpy as np
```

```python
import pandas as pd
pd.__version__
df = pd.read_csv('C:/Users/LENOVO/Desktop/PCD/tweet.txt', sep="\n",header=None)
df
```

```python
list_listword=[]#liste des listes des mots de chaque tweet
list_tweets=df.values.tolist()#liste des tweets
list_tweets
```

```python
print(list_tweets[0])
for i in range(0,len(list_tweets)):
  list_listword.append(list_tweets[i][0].split())
list_listword
```

```python
length=5+1
lines=[]#les tweets decale par 11 mots
for j in range(len(list_tweets)):
    for i in range(length,len(list_listword[j])):
        seq=list_listword[j][i-length:i]
        line=' '.join(seq)
        lines.append(line)
print(lines[0])
print(lines[1])
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(lines)
print(lines)
sequences = tokenizer.texts_to_sequences(lines)
#print(sequences)
print(len(sequences[0]))
vocab_size=len(tokenizer.word_index)
```

```python

sequences = np.array(sequences[:20])
print(sequences)

X = sequences[:, :3]
#print(X)
#print(X.shape)
#print(y)
#print(y.shape)

y =sequences[:,3:6]
#50 1 5
X=X.reshape(20, 3, 1)
y=y.reshape(20, 3, 1)

print(y)
print(y.shape)
print(X)
print(X.shape)
```

```python
from keras import backend as K
import keras.backend as kerback
from keras.models import Model
import tensorflow 
#model = Sequential()
x0=Input(shape=(3, 1))
# encoder layer
#model.add(LSTM(100, activation='relu', input_shape=(3, 1)))
x1=LSTM(100, activation='relu', input_shape=(3, 1))(x0)

# repeat vector
#model.add(RepeatVector(3))
x2=RepeatVector(3)(x1)
# decoder layer
#model.add(LSTM(100, activation='relu', return_sequences=True))
x3=LSTM(100, activation='relu', return_sequences=True)(x2)
#model.add(TimeDistributed(Dense(1)))
x4=TimeDistributed(Dense(1))(x3)

unstacked = Lambda(lambda x: tensorflow.unstack(x, axis=2))(x4)
dense_outputs = [Dense(3)(x) for x in unstacked]
merged = Lambda(lambda x: K.stack(x, axis=2))(dense_outputs)
y6=Softmax(axis=-1)(merged)
model = Model(x0, y6)
model.compile(optimizer='adam', loss='categorical_crossentropy')

print(model.summary())

```

```python
model.fit(X, y, epochs=200, validation_split=0.2, verbose=1, batch_size=3)
```

```python
#test_input = np.array([300, 305, 310])
test_input=X[6:9]
print(y[6:9])
print(test_input)
encoded = pad_sequences([test_input], maxlen = 3, truncating='pre')

result = model.predict(test_input, batch_size=3, verbose=0)
#print(test_output)
print(result)
```
